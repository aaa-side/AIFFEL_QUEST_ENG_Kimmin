```python
# Q1 - 2
def is_palindrome(word):

    return word == word[::-1]

def palindrome_classifier(targets):

    result = {
        "회문O": [],
        "회문X": []
    }

    palindrome_count = 0

    for word in targets:

        if is_palindrome(word):

            result["회문O"].append(word)
            palindrome_count += 1

        else:
            result["회문X"].append(word)

    ratio = palindrome_count / len(targets)

    return result, palindrome_count, ratio


targets = ['기러기','까마귀','토마토','스윗소스','스위스','인도인','한국인','별똥별','우영우','우향우']


print(palindrome_classifier(targets))
```

```python
#Q2 - 2
scores = {
    '유원영' : {    # 유원영 = key
        '국어' : 94,    # 국영수와 점수는 value
        '영어' : 91,
        '수학' : 89 
    },
    
    '설지민' : {
        '국어' : 88,
        '영어' : 96,
        '수학' : 90
    },
    
    '신윤아' : {
        '국어' : 86,
        '영어' : 88,
        '수학' : 77
    },
    
    '장윤아' : {
        '국어' : 98,
        '영어' : 76,
        '수학' : 92
    }   
    
}


def score_analysis(name):

    student_scores = scores[name]
    student_count = len(scores)   
    subject_count = len(student_scores) 
    
    korean_total = 0
    english_total = 0
    math_total = 0
    
    for student in scores.values():
        korean_total += student["국어"]
        english_total += student["영어"]
        math_total += student["수학"]
        
    korean_avg = korean_total / student_count
    english_avg = english_total / student_count
    math_avg = math_total / student_count
    
    student_avg = sum(student_scores.values()) / subject_count
    class_avg = (korean_avg + english_avg + math_avg) / subject_count
    
    kor_diff = student_scores['국어'] - korean_avg
    eng_diff = student_scores['영어'] - english_avg
    math_diff = student_scores['수학'] - math_avg
    total_diff = student_avg - class_avg
    
    print(f"[{name} 성적 분석]")
    print(f"국어: {student_scores['국어']}점, 학급 평균({korean_avg:.2f}점) 대비 {kor_diff:+.2f}")
    print(f"영어: {student_scores['영어']}점, 학급 평균({english_avg:.2f}점) 대비 {eng_diff:+.2f}")
    print(f"수학: {student_scores['수학']}점, 학급 평균({math_avg:.2f}점) 대비 {math_diff:+.2f}")
    print()  # 한 줄 띄우기
    print(f"전과목 평균: {student_avg:.2f}점, 학급 평균({class_avg:.2f}점) 대비 {total_diff:+.2f}")

name_input = input("이름을 입력하세요: ")
score_analysis(name_input)
```
